```python
from datetime import datetime

# Kayıtlı Kullanıcılar ve Bakiyeleri
kullanicilar = {
    "BeYZa": {"sifre": "4.6.4.6", "bakiye": 5000.0},
    "SuEda": {"sifre": "1881", "bakiye": 3500.0},
    "MelİHA": {"sifre": "GizliKod456", "bakiye": 4000.0},
    "ÖmEr": {"sifre": "NeDiMKeNGeR", "bakiye": 4500.0},
    "Endüstri": {"sifre": "HKÜ.Müh", "bakiye": 3000.0}
}

# Kullanıcı Giriş Doğrulaması
aktif_kullanici = None
while True:
    kullanici_adi = input("Kullanıcı Adı: ")
    sifre = input("Şifre: ")
    
    # Kullanıcı adı ve şifre kontrolü
    if kullanici_adi in kullanicilar and kullanicilar[kullanici_adi]["sifre"] == sifre:
        aktif_kullanici = kullanici_adi
        print(f"\nGiriş başarılı! Hoş geldiniz, {kullanici_adi}. Uygulamaya yönlendiriliyorsunuz...".upper())
        break
    else:
        print("Kullanıcı adı veya şifre yanlış! Lütfen tekrar deneyin.".upper())

# Harcama takibi uygulaması
harcamalar = []
toplam_miktar = 0  # Toplam harcamayı izlemek için ayrı bir değişken

while True:
    print(f"\nMevcut bakiye: {kullanicilar[aktif_kullanici]['bakiye']:.2f} TL")
    print("\n1. Fatura/Abonelik Ekle")
    print("2. Yaklaşan Ödeme Tarihlerini Görüntüle")
    print("3. Toplam Aylik Harcamalari Görüntüle")
    print("4. Çikiş")

    secim = input("Bir işlem seçin (1-4): ")

    if secim == "1":
        print("\n** Fatura/Abonelik Ekleme **")

        ad = input("Fatura veya Abonelik Adi: ")
        miktar = float(input("Aylik Tutar: "))

        # Bakiyeyi kontrol et
        if miktar > kullanicilar[aktif_kullanici]["bakiye"]:
            print(f"YETERSİZ BAKİYE! Mevcut bakiyeniz: {kullanicilar[aktif_kullanici]['bakiye']:.2f} TL.".upper())
            continue

        tarih = input("Son Ödeme Tarihi (yyyy-aa-gg): ")
        tarih_parcalari = tarih.split("-")

        if len(tarih_parcalari) == 3:
            yil = int(tarih_parcalari[0])
            ay = int(tarih_parcalari[1])
            gun = int(tarih_parcalari[2])

            son_odeme_tarihi = datetime(yil, ay, gun)

            # Harcama ekle ve bakiyeyi güncelle
            harcamalar.append({"ad": ad, "miktar": miktar, "tarih": son_odeme_tarihi})
            toplam_miktar += miktar
            kullanicilar[aktif_kullanici]["bakiye"] -= miktar
            print(f"{ad} başarıyla eklendi! Güncel toplam harcama: {toplam_miktar:.2f} TL")
            print(f"Kalan bakiye: {kullanicilar[aktif_kullanici]['bakiye']:.2f} TL")
        else:
            print("Hatalı tarih formatı! Lütfen yyyy-aa-gg şeklinde giriniz.")

    elif secim == "2":
        print("\n** Yaklaşan Ödeme Tarihlerini Görüntüle **")

        bugun = datetime.now()

        for harcama in harcamalar:
            kalan_gun = (harcama["tarih"] - bugun).days

            if kalan_gun <= 5 and kalan_gun >= 0:
                print(f"{harcama['ad']} - {harcama['tarih'].strftime('%Y-%m-%d')} (Kalan Gün: {kalan_gun})")
            elif kalan_gun > 5:
                print(f"{harcama['ad']} - {harcama['tarih'].strftime('%Y-%m-%d')} (Kalan Gün: {kalan_gun})")
            elif kalan_gun < 0:
                print(f"{harcama['ad']} - {harcama['tarih'].strftime('%Y-%m-%d')} (Gün Geçmiş: {abs(kalan_gun)})")

    elif secim == "3":
        print("\n** Toplam Aylik Harcamalar **")
        print(f"Toplam Aylik Harcama: {toplam_miktar:.2f} TL")

    elif secim == "4":
        print(f"Uygulama kapatılıyor. Toplam harcamalarınız: {toplam_miktar:.2f} TL".upper())
        print(f"Kalan bakiye: {kullanicilar[aktif_kullanici]['bakiye']:.2f} TL")
        break

    else:
        print("Geçersiz seçim! Lütfen 1-4 arasında bir değer girin.".upper())




```
